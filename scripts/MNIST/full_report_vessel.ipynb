```python
import pandas as pd
import numpy as np
```

```python
benchmarks = pd.DataFrame({
        "model": ["ResNet-18 + 2.5D", "ResNet-18 + 3D", "ResNet-18 + ACS", "ResNet-50 + 2.5D", "ResNet-50 + 3D", "ResNet-50 + ACS", "auto-sklearn", "AutoKeras"],
        "acc": [0.772, 0.721, 0.754, 0.763, 0.745, 0.758, 0.802, 0.705],
        "roc_auc": [0.718, 0.827, 0.839, 0.732, 0.828, 0.828, 0.828, 0.804],
        "parameters": "Unknown",
    },
    columns=["model", "acc", "roc_auc", "parameters"]
)
benchmarks
```

```python
metrics = pd.read_csv("metrics_vessel.csv")
metrics.rename(columns={"Unnamed: 0": "model"}, inplace=True)
metrics.drop(columns=["loss"], inplace=True)

metrics[["acc", "roc_auc"]] = np.round(metrics[["acc", "roc_auc"]], 3)
metrics.sort_values("acc", ascending=False, inplace=True)

metrics = pd.concat([benchmarks, metrics], axis=0)
metrics.reset_index(drop=True, inplace=True)

metrics
```

```python
metrics.to_csv("metrics_combined_vessel.csv")
```
