```python
import pathlib
import medmnist

import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
```

```python
def show(imgs):
    plt.figure(figsize=(16, 10))
    for i, image in enumerate(imgs):
        plt.subplot(4, int(len(imgs) // 4), i + 1)
        npimg = image.numpy()
        plt.imshow(npimg, interpolation='nearest')
        plt.title(f"TS {i}")
        plt.xticks([])
        plt.yticks([])

train_data = medmnist.NoduleMNIST3D(split="train", download=False)
train_dataloader = DataLoader(train_data, batch_size=4, shuffle=False)

# Display image and label.
train_features, train_labels = next(iter(train_dataloader), 100)
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
imgs = train_features[0].squeeze()
label = train_labels[0]

show(imgs)

plt.savefig(f"../../assets/datasets_nodule.png", bbox_inches='tight')
print(f"Label: {label}")
```
