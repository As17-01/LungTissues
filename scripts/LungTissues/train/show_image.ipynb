```python
import pathlib
import sys

import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader

sys.path.append("../../")

from src.datasets import RandomDataset
from src.datasets import SingleDataset
```

```python
def show(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)), interpolation='nearest')

load_dir = pathlib.Path("../../../data/intermediate")

train_data = RandomDataset(annotation_file=load_dir / "train.csv")

train_dataloader = DataLoader(train_data, batch_size=1, shuffle=True)

# Display image and label.
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0][0].squeeze()
label = train_labels[0]

show(img)
print(f"Label: {label}")
```
